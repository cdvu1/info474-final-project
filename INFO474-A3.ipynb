```python
import csv 
import pandas as pd 
import numpy as np
```

```python
data = pd.DataFrame(pd.read_csv('gun_violence_data.csv'))
```

```python
data['Date'] = pd.to_datetime(data['Date']).dt.year
data[(data.Date <= 2008) & (data.Date <= 2018)]
```

```python
test_year = data[(data['Date'] == 1998)]
```

```python
years = np.array(range(2008,2019))
```

```python
for year in years: 
    yearData = data[(data['Date'] == year)]
    totVictims = yearData['Total Injured/Killed Victims'].sum()
    print(year, totVictims)
    print(year, yearData['Shooter Age'].mean())
    print("Counts", yearData.size)
```

```python
totVictims = []
shooterAvgAge = []
totShootings = []

for year in years: 
    yearData = data[(data['Date'] == year)]
    totVictims.append(yearData['Total Injured/Killed Victims'].sum())
    shooterAvgAge.append(round(yearData['Shooter Age'].mean()))
    totShootings.append(yearData.size)
```

```python
cleanData = {'Year' : years, 'Total Shootings': totShootings, 'Total Victims': totVictims, 'Shooter Avg Age': shooterAvgAge}
cleanDf = pd.DataFrame(data=cleanData)
```

```python
cleanDf.to_csv(index=False)
```
